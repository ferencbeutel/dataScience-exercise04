```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

countriesFrame = pd.read_csv('https://raw.githubusercontent.com/edlich/eternalrepo/master/DS-WAHLFACH/countries.csv')

print("###Countries CSV###")

print("")

print("Sorted by name: ")
print(countriesFrame)

print("")

print("columns: " + ", ".join(countriesFrame.columns.base))
print("rows: " + str(int(countriesFrame.size / len(countriesFrame.columns.base))))
print("size: " + str(countriesFrame.size))

print("")

print("last 4 rows: ")
print(countriesFrame.tail(4))

print("")

print("All Countries using the EURO: ")
print(countriesFrame.loc[countriesFrame["Currency"] == "EUR"])

print("")

print("Only Name & Currency")
print(countriesFrame.filter(items=["Name", "Currency"]))

print("")

print("Countries with BIP > 2000")
print(countriesFrame.loc[countriesFrame["BIP"] > 2000])

print("")

print("Countries with more than 50 and less than 150 million people")
print(countriesFrame.loc[(countriesFrame["People"] > 5000000) & (countriesFrame["People"] < 150000000)])

print("")

print("Modify BIP column name: ")
countriesFrame.rename(columns={'BIP':'Bip'}, inplace=True)
print("columns: " + ", ".join(countriesFrame.columns.base))

print("")

print("BIP Sum: " + str(countriesFrame["Bip"].sum()) + " Billion $")

print("")

print("Average People: " + str(countriesFrame["People"].sum() / int(countriesFrame.size / len(countriesFrame.columns.base))))

print("")

print("Aliasing the Area Values: ")
countriesFrame["Area"] = np.where(countriesFrame["Area"] <= 1000000, "SMALL", "BIG")
print(countriesFrame)
```
